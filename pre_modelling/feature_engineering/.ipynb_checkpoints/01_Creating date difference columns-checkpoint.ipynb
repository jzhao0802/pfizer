<h1> Create date difference variables <h1>

```python
dict_grouped_feat = utils.generate_dict_grouped_columns(df, dict_config, list_keys = ['first_exp_date_cols', 'last_exp_date_cols'])
```

```python
dict_grouped_feat;
```

```python
df_test = pd.DataFrame()
for k, v in dict_grouped_feat.items():
    index_max_dd = k + '_INDEX_MAX_DD'
    index_min_dd = k + '_INDEX_MIN_DD'
    max_min_dd = k + '_MAX_MIN_DD'
    df_test[index_max_dd] = (df['INDEX_DATE'] - df[v['last_exp_date']]).dt.days
    df_test[index_min_dd] = (df['INDEX_DATE'] - df[v['first_exp_date']]).dt.days
    df_test[max_min_dd] = (df[v['last_exp_date']] - df[v['first_exp_date']]).dt.days
```

```python
# putting all the date columns in a list
date_cols_to_delete = [list(dicts.values()) for dicts in dict_grouped_feat.values()]
date_cols_to_delete = [item for sublist in date_cols_to_delete for item in sublist]
date_cols_to_delete.extend(['INDEX_DATE', 'LOOKBACK_DATE'])
df[date_cols_to_delete].head()
```

```python
df_no_dates = df.drop(columns = date_cols_to_delete)
df_no_dates.head()
```

```python
df_concat = pd.concat([df_no_dates, df_test], axis=1)
```

```python
ls_colnames = df_concat.columns.tolist()
map_colnames = {colname: utils.clean_string(colname) for colname in ls_colnames}
df_concat.rename(columns=map_colnames, inplace=True)
df_concat.head()
```

```python
# Grouping the columns of the dataframe together
ls_colnames = df_concat.columns.tolist()
ls_features = ls_colnames[4:]
ls_features.sort()
ls_colnames[4:] = ls_features
```

```python
df_concat = df_concat[ls_colnames]
```
