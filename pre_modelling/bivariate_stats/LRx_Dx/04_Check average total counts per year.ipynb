```python
list_colnames = df.columns.tolist()
```

```python
# Slightly higher claims per year for positive group
suffix = '_claim_cnt$'
colnames_claim_cnt = [colname for colname in list_colnames if re.search(suffix, colname)]

df_claim_cnt = df[['pn_flag'] + colnames_claim_cnt]

df_sum = pd.DataFrame({'sum':df_claim_cnt[colnames_claim_cnt].sum(axis=1)})
df_sum['pn_flag'] = df_claim_cnt['pn_flag']

df_avg_counts_per_year = df_sum.groupby('pn_flag').mean()/2

df_avg_counts_per_year
```

```python
# BUT also slightly longer average datewindow, so result of how data is cut off?
suffix = '_max_min_dd$'
colnames_max_min_dd = [colname for colname in list_colnames if re.search(suffix, colname)]

df_max_min_dd = df[['pn_flag'] + colnames_max_min_dd]

df_sum = pd.DataFrame({'avg':df_max_min_dd[colnames_max_min_dd].mean(axis=1)})
df_sum['pn_flag'] = df_max_min_dd['pn_flag']

df_avg_counts_per_year = df_sum.groupby('pn_flag').mean()

df_avg_counts_per_year
```
