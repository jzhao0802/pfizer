### Remove columns of later metastatic dates and sites

```python
ls_metac_colnames = ['MET_DATE2', 'METAC_SITE_NM2', 'MET_DATE3', 'METAC_SITE_NM3', 'MET_DATE4', 'METAC_SITE_NM4', 'MET_DATE5',
                     'METAC_SITE_NM5', 'MET_DATE6', 'METAC_SITE_NM6', 'MET_DATE7', 'METAC_SITE_NM7']
```

```python
df_onc_no_metac = df_onc.drop(columns=ls_metac_colnames)
```

```python
df_onc_no_metac.head()
```

## Make dummy variables

### METAC_SITE_NM1 

```python
# METAC_SITE_NM1 is a categorical column we want to keep
df_dummies = pd.get_dummies(df_onc_no_metac['METAC_SITE_NM1'], prefix = 'METAC_SITE')
df_dummies.head()
```

```python
# Clean up column names
dummy_colnames = [re.sub(r'&', '_', colname) for colname in df_dummies.columns]
dummy_colnames = [utils.clean_string(colname) for colname in dummy_colnames]
df_dummies.columns = dummy_colnames
df_dummies.head()
```

```python
df_onc_no_metac = df_onc_no_metac.drop(columns = ['METAC_SITE_NM1', 'MET_DATE1'])
```

```python
df_onc_no_metac.head()
```

```python
df_with_metac1 = pd.concat([df_onc_no_metac, df_dummies], axis=1)
```

### ONC_LATEST_STAGE 

```python
df_oncstage_dummies = pd.get_dummies(df_with_metac1['ONC_LATEST_STAGE'])
```

```python
clean_colnames = [utils.clean_string(colname) for colname in df_onc_dummies.columns]
df_oncstage_dummies.columns = clean_colnames
```

```python
df_with_metac_with_onc = pd.concat([df_with_metac1, df_oncstage_dummies], axis=1)
```

```python
df_with_metac_with_onc = df_with_metac_with_onc.drop(columns = ['ONC_LATEST_STAGE'])
df_with_metac_with_onc.head()
```
