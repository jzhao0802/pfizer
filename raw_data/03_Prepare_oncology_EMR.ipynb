```python
# Function to clean up newly created column names in the dummies

def clean_string(s):
    """
    Function that cleans up string, typically used for column names. Changes
    the incoming string s so that:
        - only lowercase
        - white spaces replaced by underscores
        - brackets replaced by underscores
        - multiple underscores in a row replaced by one
          underscore (e.g. ' (' -> '__' -> '_'
    :param s: string
    :return: new string
    """
    # only lowercase
    s = s.lower()
    # whites spaces, brackets, &, ;, , replaced by underscores
    s = re.sub(r'( )|(\()|(\))|(;)|(,)|(&)|(:)', '_', s)
    # multiple underscores in a row replaced by one underscore
    s = re.sub(r'(_)\1+', r'\1', s)
    # trailing underscores are removed
    s = re.sub(r'[_]+$', '', s)
    return s
```

### Remove columns of later metastatic dates and sites

```python
ls_metac_colnames = ['MET_DATE2', 'METAC_SITE_NM2', 'MET_DATE3', 'METAC_SITE_NM3', 'MET_DATE4', 'METAC_SITE_NM4', 'MET_DATE5',
                     'METAC_SITE_NM5', 'MET_DATE6', 'METAC_SITE_NM6', 'MET_DATE7', 'METAC_SITE_NM7']
```

```python
df_onc_no_metac = df_onc.drop(columns=ls_metac_colnames)
```

```python
df_onc_no_metac.head()
```

## Make dummy variables

### METAC_SITE_NM1 

```python
# METAC_SITE_NM1 is a categorical column we want to keep
df_dummies = pd.get_dummies(df_onc_no_metac['METAC_SITE_NM1'], prefix = 'METAC_SITE')
df_dummies.head()
```

```python
# Clean up column names
dummy_colnames = [clean_string(colname) for colname in df_dummies.columns]
df_dummies.columns = dummy_colnames
df_dummies.head()
```

```python
df_onc_no_metac = df_onc_no_metac.drop(columns = ['METAC_SITE_NM1', 'MET_DATE1'])
```

```python
df_onc_no_metac.head()
```

```python
df_with_metac1 = pd.concat([df_onc_no_metac, df_dummies], axis=1)
```

### ONC_LATEST_STAGE 

```python
df_oncstage_dummies = pd.get_dummies(df_with_metac1['ONC_LATEST_STAGE'])
```

```python
clean_colnames = [clean_string(colname) for colname in df_oncstage_dummies.columns]
df_oncstage_dummies.columns = clean_colnames
```

```python
df_with_metac_with_onc = pd.concat([df_with_metac1, df_oncstage_dummies], axis=1)
```

```python
df_with_metac_with_onc = df_with_metac_with_onc.drop(columns = ['ONC_LATEST_STAGE'])
df_with_metac_with_onc.head()
```

### Look for other categorical columns

```python
# Four more columns should be dummified
ls_not_numeric = [not pd.api.types.is_numeric_dtype(dtype) for dtype in df_with_metac_with_onc.dtypes]
prog = re.compile('DATE[0-9]*$')
ls_not_date = [not bool(prog.search(colname)) for colname in df_with_metac_with_onc.columns]
ls_both = [num and date for num, date in zip(ls_not_numeric, ls_not_date)]
df_with_metac_with_onc.loc[:,ls_both].nunique()
```

```python
ls_other_columns = df_with_metac_with_onc.loc[:, ls_both].columns
```

```python
df_other_dummies = pd.get_dummies(df_with_metac_with_onc[ls_other_columns])
dummy_colnames = df_other_dummies.columns
dummy_colnames = [clean_string(colname) for colname in dummy_colnames]
df_other_dummies.columns = dummy_colnames
```

```python
df = df_with_metac_with_onc.drop(columns = ls_other_columns)
df_total = pd.concat([df, df_other_dummies], axis = 1)
```

```python
all_colnames = [clean_string(colname) for colname in df_total.columns]
df_total.columns = all_colnames
df_total.head()
```

```python
df_total = df_total[ls_columns_reordered]
```
